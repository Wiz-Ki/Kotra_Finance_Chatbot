```python
# 이 셀을 실행하세요
from llm import get_rag_chain
from IPython.display import display, Markdown
import time

# 1. 질문 설정
query = '상품권 구매 시 정산은 어떻게 해야해?'

# 2. 체인 실행
print("⏳ RAG Chain을 실행 중입니다... (검색 + 답변 생성)")
start_time = time.time()

qa_chain = get_rag_chain()
response = qa_chain.invoke(
    {"input": query},
    config={"configurable": {"session_id": "debug_notebook_01"}}
)

end_time = time.time()
print(f"✅ 완료! (소요시간: {end_time - start_time:.2f}초)")

# 3. 결과 파싱
answer = response["answer"]
retrieved_docs = response["context"]

# ----------------------------------------------------------------------
# [결과 출력 1] AI 답변 (Markdown으로 깔끔하게)
# ----------------------------------------------------------------------
display(Markdown("---"))
display(Markdown(f"### ❓ 질문: **{query}**"))
display(Markdown(f"### 💡 답변:\n{answer}"))
display(Markdown("---"))

# ----------------------------------------------------------------------
# [결과 출력 2] Retrieve 된 문서 정보 (메타데이터 + 내용 미리보기)
# ----------------------------------------------------------------------
print(f"🔍 [Retrieve 된 문서 상세 확인] 총 {len(retrieved_docs)}개")
print("※ 의도: 1~3번은 '정산지침', 4~5번은 '교육자료'여야 함\n")

for i, doc in enumerate(retrieved_docs):
    # 메타데이터 추출
    meta = doc.metadata
    origin = meta.get('origin_pdf', '❌출처없음')
    page = meta.get('page_num', '❌페이지없음')
    
    # 내용 줄바꿈 제거 및 미리보기
    content_preview = doc.page_content.replace('\n', ' ')[:150] 
    
    print(f"[{i+1}순위] 📄 {origin} (Page: {page})")
    print(f"   📝 내용: {content_preview}...")
    print("-" * 60)
```
